```python
train, test = mri_levels.split_frame([0.8], seed=623)
```

```python
print("train:%d test:%d" % (train.nrows, test.nrows))
```

### feature selection 

```python
y = "SeriesDescription"
ignore = ["AccessionNumber", "BitsAllocated", "FileName", "Rows"] 
x = list(set(train.names) - set(ignore))
```

### setup automl h20 

```python
H2OAutoML(nfolds=5, max_runtime_secs=3600, max_models=None, stopping_metric='AUTO', stopping_tolerance=None, stopping_rounds=3, seed=None, project_name=None)
```

### can handle imbalance too

```python
aml = H2OAutoML(max_models=25, max_runtime_secs_per_model=30, seed=623, project_name='classification', exclude_algos=
               ['StackedEnsemble','DeepLearning'],
                balance_classes=True, class_sampling_factors=[0.5,1.25])
%time aml.train(x=x, y=y, training_frame=train)
```

## model explaination

```python
explain_model = aml.leader.explain(train)
```
